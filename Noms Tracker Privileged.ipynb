```python
import requests
from bs4 import BeautifulSoup as bs
import lxml
```

```python
r = requests.get('https://www.senate.gov/legislative/LIS/nominations/NomPrivileged.xml')
```

```python
soup = bs(r.text, "xml")
```

```python
print(soup)
```

```python
congress = soup.find('Congress').text
session = soup.find('SessionNumber').text
results = []
for nom in soup.find_all('Nomination'):
    d={}
    d["Congress"] = congress
    d["Session"] = session
    d["Nomination Display Number"] = nom.NominationDisplayNumber.text
    d["Received Date"] = nom.ReceivedDate.text
    d["Reporting Stage Date"] = nom.ReportingStageDate.text
    d["Organization"] = nom.Organization.text
    d["Reporting Description"] = nom.ReportingDescription.text
    d["Last Updated"] = nom.LastUpdated.text
    d["Senate Committee Code"] = nom.Committees.InformationRequested.Committee.SenateCommitteeCode.text
    d["Committee Full Name"] = nom.Committees.InformationRequested.Committee.CommitteeFullName.text
    results.append(d)
print(results)
    
```
