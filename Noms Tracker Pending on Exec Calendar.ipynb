```python
import requests
from bs4 import BeautifulSoup as bs
import lxml
```

```python
r = requests.get('https://www.senate.gov/legislative/LIS/nominations/NomCivilianPendingCalendar.xml')
```

```python
soup = bs(r.text, "xml")
```

```python
print(soup)
```

```python
congress = soup.find('Congress').text
session = soup.find('SessionNumber').text
results = []
for nom in soup.find_all('Nomination'):
    d={}
    d["Congress"] = congress
    d["SessionNumber"] = session
    d["Nomination Display Number"] = nom.NominationDisplayNumber.text
    d["Received Date"] = nom.ReceivedDate.text
    d["Reporting Stage Date"] = nom.ReportingStageDate.text
    d["Organization"] = nom.Organization.text
    d["Reporting Description"] = nom.ReportingDescription.text
    d["Executive Calendar Number"] = nom.ExecutiveCalendarNumber.text
    d["Committee"] = nom.Committees.text
    d["Committee Code"] = nom.Committees.CommitteeReferrals.Committee.SenateCommitteeCode.text
    results.append(d)
print(results)
```

```python
congress = soup.find('Congress').text
session = soup.find('SessionNumber').text
committees = soup.find('Committees').text
results = []
for nom in soup.find_all('Nomination'):
    d={}
    d["Committee"] = committees
    for com in soup.find_all('CommitteeReferrals'):
        d={}
        d["Committee Referral Date"] = com.CommitteeReferralDate.text
        d["Commitee Reported Out Date"] = com.CommitteeReportedOutDate.text
        d["Senate Committee Code"] = com.Committee.SenateCommitteeCode.text
        d["Committee Name"] = com.Committee.CommitteeFullName.text
    results.append(d)
print(results)
```
